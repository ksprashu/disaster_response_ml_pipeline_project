```python
def tokenize_nonstop(text):
    tokens = word_tokenize(text)
    tokens = [tok for tok in tokens if re.search('[A-Za-z0-9]', tok) != None]
    
    return tokens


def tokenize_stop(text):
    tokens = word_tokenize(text)
    tokens = [tok for tok in tokens if re.search('[A-Za-z0-9]', tok) != None]
    tokens = [tok for tok in tokens if tok not in stopwords.words('english')]
    
    return tokens


def tokenize_lemmatize(text):
    tokens = word_tokenize(text)
    tokens = [tok for tok in tokens if re.search('[A-Za-z0-9]', tok) != None]
    tokens = [tok for tok in tokens if tok not in stopwords.words('english')]
    tokens = [wnl.lemmatize(wnl.lemmatize(tok), pos='v') for tok in tokens]            
    
    return tokens


def tokenize_nonstop_lemmatize(text):
    tokens = word_tokenize(text)
    tokens = [tok for tok in tokens if re.search('[A-Za-z0-9]', tok) != None]
    tokens = [wnl.lemmatize(wnl.lemmatize(tok), pos='v') for tok in tokens]            
    
    return tokens
```
